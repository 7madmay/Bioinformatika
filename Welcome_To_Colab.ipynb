```
# input jumlah makanan
jumlah = int(input("Masukkan jumlah makanan yang dikonsumsi hari ini: "))

total_kalori = 0

# looping proses input kalori tiap makanan
for i in range(1, jumlah + 1):
    kalori = float(input(f"Masukkan kalori makanan ke-{i}: "))
    total_kalori += kalori

# tampilkan total kalori
print("\nTotal kalori harian =", total_kalori, "kkal")

# seleksi kondisi status nutrisi
if total_kalori < 1500:
    print("Status: Kekurangan kalori")
elif 1500 <= total_kalori <= 2500:
    print("Status: Normal / Seimbang")
elif 2500 < total_kalori <= 3000:
    print("Status: Kelebihan ringan")
else:
    print("Status: Kelebihan berat (berisiko)")
```

```
while True:
    nama_list = []
    nilai_array = []
    data = {}
    jumlah = int(input("Masukkan jumlah mahasiswa: "))

    for i in range(jumlah):
        nama = input("Nama: ")
        nilai = float(input("Nilai: "))

        nama_list.append(nama)
        nilai_array.append(nilai)
        data[nama] = nilai

    print("\nData Mahasiswa:")
    for n in data:
        print(n, "=", data[n])

    rata = sum(nilai_array) / jumlah
    print("Rata-rata =", rata)

    print("\nStatus:")
    for n in data:
        if data[n] >= 75:
            print(n, "Lulus")
        else:
            print(n, "Tidak Lulus")

    if input("\nUlang? (y/t): ") == "t":
        break
```

```
while True:
    lokasi_list = []
    data = {}
    kategori = ("Aman", "Tercemar Ringan", "Tercemar Sedang", "Tercemar Berat")
    jumlah = int(input("Masukkan jumlah lokasi sungai: "))

    for i in range(jumlah):
        lokasi = input(f"Masukkan nama lokasi ke-{i+1}: ")
        polutan = float(input(f"Masukkan konsentrasi polutan (mg/L) di {lokasi}: "))

        lokasi_list.append(lokasi)
        data[lokasi] = polutan

    print("\nData Ekotoksikologi:")
    for l in data:
        print(l, "=", data[l], "mg/L")

    total = sum(data.values())
    rata = total / jumlah
    print("\nRata-rata polutan =", rata, "mg/L")

    # mencari nilai tertinggi dan terendah
    max_lokasi = max(data, key=data.get)
    min_lokasi = min(data, key=data.get)

    print("Lokasi dengan polutan tertinggi:", max_lokasi, "=", data[max_lokasi])
    print("Lokasi dengan polutan terendah:", min_lokasi, "=", data[min_lokasi])

    print("\nKlasifikasi Pencemaran:")
    for l in data:
        nilai = data[l]

        if nilai <= 1:
            status = kategori[0]
        elif 1 < nilai <= 5:
            status = kategori[1]
        elif 5 < nilai <= 10:
            status = kategori[2]
        else:
            status = kategori[3]

        print(l, ":", status)

    ulang = input("\nIngin mengulang program? (ya/tidak): ").lower()
    if ulang == "tidak":
        print("Program selesai")
        break
```
