```
nama_barang = input("Masukkan nama barang: ")
harga = float(input("Masukkan harga barang: "))
jumlah = int(input("Masukkan jumlah beli: "))

total = harga * jumlah
if total > 100000:
    diskon = 0.1 * total
else:
    diskon = 0
bayar = total - diskon

print(f"Total: {total:.0f}")
print(f"Diskon: {diskon:.0f}")
print(f"Total yang harus dibayar: {bayar:.0f}")
```

```
# input jumlah makanan
jumlah = int(input("Masukkan jumlah makanan yang dikonsumsi hari ini: "))

total_kalori = 0

# looping proses input kalori tiap makanan
for i in range(1, jumlah + 1):
    kalori = float(input(f"Masukkan kalori makanan ke-{i}: "))
    total_kalori += kalori

# tampilkan total kalori
print("\nTotal kalori harian =", total_kalori, "kkal")

# seleksi kondisi status nutrisi
if total_kalori < 1500:
    print("Status: Kekurangan kalori")
elif 1500 <= total_kalori <= 2500:
    print("Status: Normal / Seimbang")
elif 2500 < total_kalori <= 3000:
    print("Status: Kelebihan ringan")
else:
    print("Status: Kelebihan berat (berisiko)")
```
